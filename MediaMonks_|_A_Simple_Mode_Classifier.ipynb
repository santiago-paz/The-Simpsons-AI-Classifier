```python
path = Path('the_simpsons')
characters = 'bart', 'homer', 'lisa', 'marge', 'maggie'

for character in characters:
    # Descarga imágenes y las guarda en la carpeta especificada:
    duckduckgo_search(path, character, f"{character} simpson", 200)
```

```python
# Limpieza manual 🧹 de los datos:
from jmd_imagescraper.imagecleaner import *
display_image_cleaner(path)
```

```python
# Trancas...
# Creamos un "Datablock". Un almacenador de datos, donde ya 
# moldeamos los datos para que el modelo los pueda entender.

data_characters = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2,seed=43),
    get_y=parent_label,
    item_tfms=Resize(128),
    batch_tfms=aug_transforms(mult=2.0))

data_characters = data_characters.new(
        item_tfms=RandomResizedCrop(224, min_scale=0.5),
        batch_tfms=aug_transforms())
```

```python
# Le pasamos la carpeta donde estan las imagenes.
dls = data_characters.dataloaders(path)

# Finalmente, instanciamos nuestro modelo (resnet18) y luego
# lo entrenamos, en 4
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```
